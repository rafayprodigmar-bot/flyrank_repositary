## 3. Quick look at the data (2-3 real numbers)

We are loading the anonymized dataset data/raw/content_refresh_anonymized.csv. By analyzing the dataset size and the actual proportion of declining pages (where trend_direction == 'down'), we can establish our exact target baseline and verify the scale of the problem.

```python
# This cell is for CODE (numbers, a query, a check).
# Write your text answer in the cell ABOVE this one — typing sentences here breaks Run All.
import pandas as pd
import os

# Path to the bundled dataset
data_path = "data/raw/content_refresh_anonymized.csv"

# If running in Colab, we might need to adjust the path or clone the repo first
if not os.path.exists(data_path):
    # Fallback for Google Colab environment pointing to your repo
    !git clone https://github.com/flyrank-bih/flyrank-ml-internship-starter.git temp_repo
    !mv temp_repo/data .
    !rm -rf temp_repo

# Load and check the data
df = pd.read_csv(data_path)
total_rows = len(df)
declining_pages = len(df[df['trend_direction'] == 'down'])
decline_percentage = (declining_pages / total_rows) * 100

print(f"Total rows in dataset: {total_rows}")
print(f"Number of declining pages: {declining_pages}")
print(f"Percentage of declining pages: {decline_percentage:.2f}%")
```
